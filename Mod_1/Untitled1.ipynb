```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
titles = soup.find_all('td', class_='a-text-left mojo-field-type-release mojo-cell-wide')
```

```python
titles[0].select('a')[0].string
```

```python

movie_title = []
for i in titles:
       movie_title.append(i.select('a')[0].string) 
```

```python
movie_title   
```

```python
release = soup.find_all('td', class_='a-text-left mojo-field-type-date a-nowrap')
```

```python
date = []
for release_date in release:
    date.append(release_date.string)
```
